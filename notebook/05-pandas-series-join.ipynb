```python
import pandas as pd

# load data ke dataframe (df)
df = pd.read_csv("data-kebakaran.csv")
df.head()
```

## _OUTER JOIN_

_OUTER JOIN_ adalah proses penggabungan 2 DataFrame sehingga hasilnya adalah DataFrame yang berisi data dari kedua DataFrame.

image.png

```python
s1 = df[df.bulan==1].groupby('penyebab').korban_jiwa.sum()
s1
```

```python
s2 = df[df.bulan==2].groupby('penyebab').korban_jiwa.sum()
s2
```

operasi _OUTER JOIN_ dilakukan lewat perintah `concat`

```python
df1 = pd.concat([df1,df2], axis=1)

# beri nama kolom
# ref: https://www.geeksforgeeks.org/add-column-names-to-dataframe-in-pandas/

# data yang kosong, isi dengan angka 0
# ref: https://datatofish.com/replace-nan-values-with-zeros/

# tampilkan data
df1
```

## INNER JOIN

_INNER JOIN_ adalah proses penggabungan 2 DataFrame sehingga hasilnya adalah DataFrame yang hanya berisi data dari DataFrame pertama (kiri) yang memiliki pasangan dengan DataFrame kedua (kanan).

image.png

```python
s1 = df[df.bulan==1].groupby('penyebab').korban_jiwa.sum()
s1
```

```python
s2 = df[df.bulan==2].groupby('penyebab').korban_jiwa.sum()
s2
```

```python

df2 = pd.merge(s1, s2, left_index=True, right_index=True)
df2
```

```python
s1 = df[df.bulan==1].groupby('sumber_info').korban_jiwa.sum()
s1.name = "jan"

s2 = df[df.bulan==2].groupby('sumber_info').korban_jiwa.sum()
s2.name = "feb"

df2 = pd.merge(s1, s2, left_index=True, right_index=True)
df2
```

## _LEFT JOIN_

_LEFT JOIN_ adalah proses penggabungan 2 DataFrame sehingga hasilnya adalah DataFrame yang hanya berisi semua data dari DataFrame pertama (kiri). Jika ia memiliki pasangan di DataFrame kedua (kanan), maka data itu akan disalin, jika tidak ada maka akan diberikan nilai _NaN (not a number)_.

image.png

Untuk melakukkan _LEFT JOIN_ kita menggunakan fungsi `join`. Fungsi `join` adalah fungsi yang menjadi bagian dari DataFrame, jadi pastikan objek yang akan di proses merupakan DataFrame.

```python
# Hasil group by harus di konversi ke DataFrame
df1 = pd.DataFrame(df[df.bulan==1].groupby('penyebab').korban_jiwa.sum())

# beri kolom data dengan nama "jan"
df1.columns = ['jan']
df1
```

```python
# Hasil group by harus di konversi ke DataFrame
df2 = pd.DataFrame(df[df.bulan==2].groupby('penyebab').korban_jiwa.sum())

# beri kolom data dengan nama "feb"
df2.columns = ['feb']
df2
```

```python
# join df1 dengan df2
df1.join(df2)
```
